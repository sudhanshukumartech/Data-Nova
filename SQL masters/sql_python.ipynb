```python
QUERY="""Select c.customer , p.product , count(f.product_code) as Count
From fact_sales f
join dim_customer c on c.customer_code=f.customer_code
join dim_product p on p.`Product code`=f.Product_code
group by 1,2
having count(f.product_code)>1
order by count(f.product_code) desc;"""
```

```python
cursor.execute(QUERY)
cursor.fetchall()
```

```python
### CREATE TABLE STUDENT

QUERY = """ DROP table if exists student;

        create table student (
        roll_number int primary key,
        student_name varchar(50),
        student_address varchar(100),
        course varchar(100));"""
```

```python
for i in QUERY.split(';'):
    if i.strip():
      cursor.execute(i)



connection.commit()
        
         
```
